```python
import re
import nltk
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
nltk.download('punkt')
```

```python
with open('pride_and_prejudice_text.txt', encoding = 'utf-8-sig') as f:
    text = f.read()
```

```python
# take care of \n characters
text = re.sub(r'\s+', ' ', text.replace('\n', ' '))
```

```python
# split text into chapters; 61 chapters total
chapters = re.split(r'CHAPTER\s\w+.\s', text)[1:]
chapters = [s.strip() for s in chapters]
```

```python
chapters[0]
```

```python
# sentences is a list of all sentences in the book
sentences = []
for i in chapters:
    chapter_sentences = nltk.sent_tokenize(i)
    for s in chapter_sentences:
        sentences.append(s)
```

## Sentiment per chapter

```python
sia = SentimentIntensityAnalyzer()

# get sentiment score for each chapter
sentiment_scores = np.zeros(len(chapters))

# calculates score for each sentence in a chapter and averages them
for i, chapter in enumerate(chapters):
    chapter_sentences = nltk.sent_tokenize(chapter)
#     sentiment_scores.append(0)
    sentiment = 0
    for s in chapter_sentences:
        sentiment += sia.polarity_scores(s)['compound']
    sentiment /= len(chapter_sentences)
    sentiment_scores[i] = sentiment
```

```python
sentiment_scores
```

```python
plt.title('sentiment scores for each chapter')
plt.xlabel('chapter')
plt.ylabel('sentiment score')
plot_x = range(len(chapters))
plot_y = sentiment_scores
plt.plot(plot_x, plot_y)
plt.show()
```

```python
sentiment_df = pd.DataFrame(sentiment_scores, columns = ['sentiment_score'])
sentiment_df['chapter'] = np.arange(1, 62)
```

## Character interactions

```python
characters = [['Elizabeth Bennet', 'Elizabeth', 'Lizzie', 'Lizzy'],
             ['Jane Bennet', 'Jane', 'Ms. Bennet'],
             ['George Wickham', 'Wickham'],
             ['Mr. Darcy', 'Darcy'],
             ['Mr. Bingley', 'Bingley'],
             ['Charlotte Lucas', 'Charlotte', 'Ms. Lucas', 'Mrs. Collins'],
             ['Lady Catherine', 'De Bourgh'],
             ['Mr. Collins', 'Collins']]
```

```python
# find subsets of 7 sentences where the provided characters appear or are mentioned
subsets_dict = {}

for char in characters:
    subsets_dict[char[0]] = []
    
for i, sentence in enumerate(sentences):
    for char in characters:
        if any(name in sentence for name in char) or any(name.upper() in sentence for name in char):
            subsets_dict[char[0]].append(" ".join([sentences[j] for j in range(max(0, i-3), min(i+3, len(sentences)-1))]))
            break
```

```python
# find percentage of sentences that mention two characters in the same subset
interactions = {'nodes': [], 'edges': []}

for i in range(len(characters)):
    if i < len(characters) - 1:
        for j in range(i+1, len(characters)):
            interactions['edges'].append({'source': characters[i][0], 'target': characters[j][0], 'weight': []})

for source_char, subset in subsets_dict.items():
    occurrences = len(subset)
    for target_char in characters:
        if source_char != target_char[0]:
            counter = 0
            for s in subset:
                if any(name in s for name in target_char) or any(name.upper() in s for name in target_char):
                    counter += 1
            w = counter / float(occurrences)
            edge = next(item for item in interactions['edges'] if (item['source'] == source_char and item['target'] == target_char[0]) or (item['target'] == source_char and item['source'] == target_char[0]))
            edge['weight'].append(w)
            # interactions['edges'].append({'source': source_char, 'target': target_char[0], 'weight': w})

for edge in interactions['edges']:
    edge['weight'] = (edge['weight'][0] + edge['weight'][1]) / 2
```

```python
interactions['edges']
```

```python
# put source and target index into edges
char_index_dict = {'Elizabeth Bennet': 0, 'Jane Bennet': 1, 'George Wickham': 2, 'Mr. Darcy': 3, 'Mr. Bingley': 4,
                   'Charlotte Lucas': 5, 'Lady Catherine': 6, 'Mr. Collins': 7}

for edge in interactions['edges']:
    edge['source_index'] = char_index_dict[edge['source']]
    edge['target_index'] = char_index_dict[edge['target']]
```

```python
# get the total weight for each character
for char in characters:
    interactions['nodes'].append({'name': char[0], 'weight': 0})
    
for edge in interactions['edges']:
    next(item for item in interactions['nodes'] if item['name'] == edge['source'])['weight'] += edge['weight']
    next(item for item in interactions['nodes'] if item['name'] == edge['target'])['weight'] += edge['weight']
```

## Character emotion intensity by event

```python
event_chapter_dict = {
    'The Beginning': (1, 'social'),
    'Meryton Ball': (3, 'social'),
    'Stay at Netherfield': (8, 'social'),
    'Netherfield Ball': (18, 'social'),
    "Mr. Collins' Proposal": (19, 'confrontation'),
    'See Mr. Darcy in Kent': (30, 'romantic'),
    "Mr. Darcy's Proposal": (34, 'romantic'),
    "Mr. Darcy's Letter": (35, 'confrontation'),
    'A Visit to Derbyshire': (43, 'romantic'),
    "Lydia's Run Away": (46, 'confrontation'),
    "Finds Out That Mr. Darcy Saved Lydia": (52, 'romantic'),
    "Mr. Bingley's Proposal": (55, 'romantic'),
    "Lady Catherine Visits": (56, 'confrontation'),
    "A New Declaration of Love": (58, 'romantic')
}

relevant_chars = [
    ['Elizabeth Bennet', 'Elizabeth', 'Lizzie', 'Lizzy'],
    ['Jane Bennet', 'Jane', 'Ms. Bennet'],
    ['Mr. Darcy', 'Darcy'],
    ['Mr. Bingley', 'Bingley']
]

intensities = {'character': [], 'main_event': [], 'intensity_score': [], 'event_type': []}
```

```python
for (event_name, (chapter, event_type)) in event_chapter_dict.items():
    chapter_sentences = nltk.sent_tokenize(chapters[chapter - 1])
    for char_array in relevant_chars:
        char = char_array[1]
        sentiment_aggregate = 0
        num_sentences = 0
        for sentence in chapter_sentences:
            if any(name in sentence for name in char_array) or any(name.upper() in sentence for name in char_array):
                sentiment_aggregate += sia.polarity_scores(sentence)['compound']
                num_sentences += 1
        if num_sentences != 0:
            intensities['character'].append(char)
            intensities['main_event'].append(event_name)
            intensities['intensity_score'].append(round(sentiment_aggregate / num_sentences, 2))
            intensities['event_type'].append(event_type)
```

```python
# desired output: intensities = {'character': [], 'main_event': [], 'intensity_score': [], 'event_type': []}
intensity_df = pd.DataFrame(intensities)
intensity_df
```

## Output

```python
# output sentiment_scores to csv
sentiment_df.to_csv('sentiment_per_chapter.csv', index = False)
# output character relations to json
with open("character_relationships.json","w") as f:
    json.dump(interactions, f)
# output character intensity to csv
intensity_df.to_csv('character_event_intensity.csv', index = False)
```
